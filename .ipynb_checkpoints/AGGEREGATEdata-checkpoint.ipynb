```python
warehouse  = pd.read_csv(r"Datasets\ts-warehouse-data.csv",sep=',')
warehouse_new=warehouse.drop(['address','status','latitude','longitude'],axis=1)
```

```python
serial = warehouse_new[warehouse_new.district == "No Data"]['s.no']
print(warehouse_new.shape)
warehouse_new = warehouse_new.drop(list(serial),inplace=True)
print(warehouse_new.shape)
```
