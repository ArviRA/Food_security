```python
districts = []
i = 0
```

```python
import os
filenames = os.listdir("Datasets/districts_wise/")
```

```python
for file in filenames:
    districts.append(pd.read_csv("Datasets/districts_wise/"+file,sep=","))
    
```

```python
for i in range(len(districts)):
    districts[i] = districts[i].groupby("crop")
    temp.append(list(districts[i].groups))
```

```python
du = set()
for i in temp:
    for j in i:
        du.add(j)
du    
    
```

```python
for i in range(len(districts)):
    districts[i] = districts[i].groupby("crop").mean()
    #print(districts[i].head(5))
    print(districts[i].shape)
    i = i+1
```

```python
crop_variety = []
for i in range(len(districts)):
    temp = []
    temp.append(filenames[i].split("_")[0])
    temp.append(districts[i].shape[0])
    crop_variety.append(temp)
    print(temp)
crop_variety = pd.DataFrame(crop_variety,columns=["district","crop_variety"])
crop_variety.to_csv('crop_diversity.csv') 
crop_variety.head()
```
