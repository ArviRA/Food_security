```
warehouse  = pd.read_csv(r"ts-warehouse-data.csv",sep=',')
warehouse_new=warehouse.drop(['address','status','latitude','longitude'],axis=1)
```

```
serial = list(warehouse_new[warehouse_new.district == "No Data"]['s.no'])
for i in range(0,len(serial)):
    serial[i]-=1
print(warehouse_new.shape,serial)
warehouse_new = warehouse_new.drop(list(serial))
print(warehouse_new.shape)
warehouse_new.head()
```

```
dis = list(warehouse_new.district.unique())
warehouse_new = warehouse_new.groupby("district").sum().drop(["s.no"],axis=1)
warehouse_new["districts"] = diss
```

```
warehouse_new
```

```
c = []
o = []
v = []
for i in range(main.shape[0]):
  l = list(warehouse_new[warehouse_new.districts == diss[i]]["capacity"])
  
  #print(diss[i],"\n",main[main.Districts == diss[i]])
  if l==[]:
    c.append(None)
  else:
    c.append(l[0])  
  
  l = list(warehouse_new[warehouse_new.districts == diss[i]]["occupancy"])
  
  #print(diss[i],"\n",main[main.Districts == diss[i]])
  if l==[]:
    o.append(None)
  else:
    o.append(l[0]) 

  l = list(warehouse_new[warehouse_new.districts == diss[i]]["vacancy"])
  
  #print(diss[i],"\n",main[main.Districts == diss[i]])
  if l==[]:
    v.append(None)
  else:
    v.append(l[0]) 
    
    
print(c,o,v,literacy.shape)
```

```
main["capacity_warehouse"] = c
main["occupancy_warehouse"] = o
main["vacancy_warehouse"] = v
```
